```python
#2 Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його випадковими дійсними числами від 0 до 1 та виведіть його значення.
import random

# Створюємо пусту матрицю розміром 3x3
matrix = [[0.0 for _ in range(3)] for _ in range(3)]

# Заповнюємо матрицю випадковими дійсними числами від 0 до 1
for i in range(3):
    for j in range(3):
        matrix[i][j] = random.uniform(0, 1)

# Виводимо значення матриці
for row in matrix:
    print(row)
```

```python
#4 Створіть масив розміром 4x4, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть мінімальний елемент.
import random

# Створюємо пусту матрицю розміром 4x4
matrix = [[0.0 for _ in range(4)] for _ in range(4)]

# Заповнюємо матрицю випадковими дійсними числами від 0 до 1
for i in range(4):
    for j in range(4):
        matrix[i][j] = random.uniform(0, 1)

# Виводимо значення матриці
for row in matrix:
    print(row)

# Знаходимо мінімальний елемент
min_element = matrix[0][0]
for i in range(4):
    for j in range(4):
        if matrix[i][j] < min_element:
            min_element = matrix[i][j]

print("Мінімальний елемент в матриці: ", min_element)
```

```python
#5 Створіть два одновимірних масиви розміром 7, заповніть їх випадковими цілими числами в діапазоні від 1 до 15 та виконайте на них поелементні операції додавання, віднімання та множення.
import random

# Створюємо два пусті одновимірні масиви розміром 7
array1 = []
array2 = []

# Заповнюємо масиви випадковими цілими числами від 1 до 15
for _ in range(7):
    array1.append(random.randint(1, 15))
    array2.append(random.randint(1, 15))

# Виводимо значення двох масивів
print("Масив 1:", array1)
print("Масив 2:", array2)

# Виконуємо поелементне додавання
sum_array = [a + b for a, b in zip(array1, array2)]
print("Результат додавання:", sum_array)

# Виконуємо поелементне віднімання
subtract_array = [a - b for a, b in zip(array1, array2)]
print("Результат віднімання:", subtract_array)

# Виконуємо поелементне множення
multiply_array = [a * b for a, b in zip(array1, array2)]
print("Результат множення:", multiply_array)
```

```python
#6 Створіть два вектори розміром 7, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток.
import random

# Створюємо два пусті вектори розміром 7
vector1 = []
vector2 = []

# Заповнюємо вектори довільними цілими числами
for _ in range(7):
    vector1.append(random.randint(1, 10))
    vector2.append(random.randint(1, 10))

# Виводимо значення двох векторів
print("Вектор 1:", vector1)
print("Вектор 2:", vector2)

# Обчислюємо скалярний добуток векторів
scalar_product = sum(a * b for a, b in zip(vector1, vector2))
print("Скалярний добуток векторів:", scalar_product)
```

```python
#7 Створіть дві матриці розміром 4x2 та 2x5, заповніть їх випадковими цілими числами в діапазоні від 1 до 12 та перемножте їх між собою.
import random

# Створюємо пусту першу матрицю розміром 4x2 і заповнюємо її випадковими цілими числами від 1 до 12
matrix1 = [[random.randint(1, 12) for _ in range(2)] for _ in range(4)]

# Виводимо значення першої матриці
print("Перша матриця:")
for row in matrix1:
    print(row)

# Створюємо пусту другу матрицю розміром 2x5 і заповнюємо її випадковими цілими числами від 1 до 12
matrix2 = [[random.randint(1, 12) for _ in range(5)] for _ in range(2)]

# Виводимо значення другої матриці
print("\nДруга матриця:")
for row in matrix2:
    print(row)

# Створюємо пусту матрицю-результат розміром 4x5
result_matrix = [[0 for _ in range(5)] for _ in range(4)]

# Перемножуємо матриці
for i in range(4):
    for j in range(5):
        for k in range(2):
            result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

# Виводимо значення матриці-результату
print("\nРезультат перемноження матриць:")
for row in result_matrix:
    print(row)
```

```python
#8 Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 8 та знайдіть її обернену матрицю.
import numpy as np

# Створюємо матрицю 2x2 з випадковими цілими числами від 1 до 8
matrix = np.random.randint(1, 9, (2, 2))

# Виводимо вихідну матрицю
print("Вихідна матриця:")
print(matrix)

# Обчислюємо обернену матрицю
inverse_matrix = np.linalg.inv(matrix)

# Виводимо обернену матрицю
print("\nОбернена матриця:")
print(inverse_matrix)
```

```python
#9 Створіть матрицю розміром 2x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її
import numpy as np

# Створюємо матрицю 2x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.uniform(0, 1, (2, 3))

# Виводимо вихідну матрицю
print("Вихідна матриця:")
print(matrix)

# Транспонуємо матрицю
transposed_matrix = matrix.T

# Виводимо транспоновану матрицю
print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
#10 Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10, обчисліть добуток кожного стовпця матриці на відповідний елемент вектору та знайдіть суму цих добутків.
import numpy as np

# Створюємо матрицю 3x4 та вектор розміром 4 з випадковими цілими числами від 1 до 10
matrix = np.random.randint(1, 11, (3, 4))
vector = np.random.randint(1, 11, 4)

# Виводимо значення матриці і вектора
print("Матриця:")
print(matrix)
print("\nВектор:")
print(vector)

# Обчислюємо добутки кожного стовпця матриці на відповідний елемент вектору
column_products = np.sum(matrix * vector, axis=0)

# Виводимо добутки
print("\nДобутки кожного стовпця:")
print(column_products)

# Знаходимо суму добутків
sum_of_column_products = np.sum(column_products)

# Виводимо суму добутків
print("\nСума добутків кожного стовпця:", sum_of_column_products)
```

```python
#12 Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток першої матриці на другу.
import numpy as np

# Створюємо першу матрицю 4x4 з випадковими цілими числами від 1 до 10
matrix1 = np.random.randint(1, 11, (4, 4))

# Створюємо другу матрицю 4x4 з випадковими цілими числами від 1 до 10
matrix2 = np.random.randint(1, 11, (4, 4))

# Виводимо першу матрицю
print("Перша матриця:")
print(matrix1)

# Виводимо другу матрицю
print("\nДруга матриця:")
print(matrix2)

# Обчислюємо добуток першої матриці на другу
product_matrix = np.dot(matrix1, matrix2)

# Виводимо результат добутку
print("\nРезультат добутку:")
print(product_matrix)
```

```python
#13 Створіть дві матриці розміром 4x4, заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть їх добуток
import numpy as np

# Створюємо першу матрицю 4x4 з випадковими дійсними числами від -1 до 1
matrix1 = np.random.uniform(-1, 1, (4, 4))

# Створюємо другу матрицю 4x4 з випадковими дійсними числами від -1 до 1
matrix2 = np.random.uniform(-1, 1, (4, 4))

# Виводимо першу матрицю
print("Перша матриця:")
print(matrix1)

# Виводимо другу матрицю
print("\nДруга матриця:")
print(matrix2)

# Обчислюємо добуток першої матриці на другу
product_matrix = np.dot(matrix1, matrix2)

# Виводимо результат добутку
print("\nРезультат добутку:")
print(product_matrix)
```

```python
#14 Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть суму елементів матриці.
import numpy as np

# Створюємо матрицю 7x7 з випадковими дійсними числами від -1 до 1
matrix = np.random.uniform(-1, 1, (7, 7))

# Виводимо матрицю (опціонально)
print("Матриця:")
print(matrix)

# Обчислюємо суму елементів матриці
sum_of_elements = np.sum(matrix)

# Виводимо суму
print("\nСума елементів матриці:", sum_of_elements)
```

```python
#15 Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
import numpy as np

# Створюємо матрицю 5x5 з випадковими дійсними числами від -0.5 до 0.5
matrix = np.random.uniform(-0.5, 0.5, (5, 5))

# Виводимо матрицю (опціонально)
print("Матриця:")
print(matrix)

# Обчислюємо вектор-стовпчик суми елементів кожного рядка матриці
row_sums = np.sum(matrix, axis=1, keepdims=True)

# Виводимо вектор-стовпчик сум
print("\nВектор-стовпчик сум рядків матриці:")
print(row_sums)
```
